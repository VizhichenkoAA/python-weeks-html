Почасовые данные после normalize: пропуски, распределения, связь ветра и осадков.

```python
from pathlib import Path
import pandas as pd

ROOT = Path('..').resolve()
norm_dir = ROOT / 'data' / 'normalized' / 'variant_03'
files = sorted(norm_dir.glob('*.csv'))
assert files, f'Нет CSV в {norm_dir}. Сначала: scripts\\run_normalize.bat'
path = files[-1]
df = pd.read_csv(path, parse_dates=['ts'])
print('file:', path.name)
df.head()
```

```python
print('shape:', df.shape)
print(df.dtypes)
df.isna().sum()
```

```python
desc = df[['temperature_2m', 'relative_humidity_2m', 'precipitation', 'wind_speed_10m']].describe()
desc
```

```python
daily = df.assign(date=df['ts'].dt.date).groupby('date').agg(
    T_mean=('temperature_2m', 'mean'),
    P_sum=('precipitation', 'sum'),
    rainy_hours=('precipitation', lambda s: (s.fillna(0) > 0).sum()),
    wind_max=('wind_speed_10m', 'max'),
)
daily.head(10)
```

```python
corr = daily[['T_mean', 'P_sum', 'wind_max', 'rainy_hours']].corr()
corr
```
