# Week 7 — визуализация mart (вариант 03, Новосибирск)

Три вопроса к данным:
1. **Время:** как меняется средняя температура по дням?
2. **Распределение:** как распределена максимальная скорость ветра?
3. **Ранжирование:** какие дни лидируют по ветру?

```python
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
from sem2_de.config import mart_dir

mart_file = sorted(mart_dir().glob('mart_daily_*.csv'))[-1]
df = pd.read_csv(mart_file, parse_dates=['date']).sort_values('date')
print(mart_file.name, df.shape)
df.head()
```

```python
plt.figure(figsize=(9,4))
plt.plot(df['date'], df['T_mean'], marker='o')
plt.title('Средняя температура по дням')
plt.xlabel('Дата')
plt.ylabel('T_mean, °C')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.savefig('docs/figures/week7_timeseries.png', dpi=120)
plt.show()
```

**Вывод 1:** На графике видна дневная динамика `T_mean`; пики и провалы соответствуют конкретным датам периода 2025-01-01…14.

```python
plt.figure(figsize=(7,4))
plt.hist(df['wind_max'], bins=8, edgecolor='white')
plt.title('Распределение wind_max')
plt.xlabel('wind_max, км/ч')
plt.ylabel('Число дней')
plt.tight_layout()
plt.savefig('docs/figures/week7_distribution.png', dpi=120)
plt.show()
```

**Вывод 2:** Большинство дней имеют умеренный ветер; хвост распределения — отдельные ветреные дни (см. ranking).

```python
top = df.nsmallest(5, 'wind_rank').sort_values('wind_max', ascending=True)
plt.figure(figsize=(8,4))
plt.barh(top['date'].dt.strftime('%Y-%m-%d'), top['wind_max'])
plt.title('Топ дней по wind_max')
plt.xlabel('км/ч')
plt.tight_layout()
plt.savefig('docs/figures/week7_ranking.png', dpi=120)
plt.show()
```

**Вывод 3:** Лидеры по ветру — 2025-01-02 и 2025-01-03 (`wind_max` ≈ 26 км/ч); это согласуется с KPI `wind_rank` в витрине.
