# Week 13 — аномалии T_mean (вариант B, Z-score)

Без target: ищем подозрительные дни по отклонению дневной средней температуры.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sem2_de.config import mart_dir

df = pd.read_csv(sorted(mart_dir().glob('mart_daily_*.csv'))[-1], parse_dates=['date'])
df = df.sort_values('date')
mu, sigma = df['T_mean'].mean(), df['T_mean'].std(ddof=0)
df['z'] = (df['T_mean'] - mu) / sigma
df['is_anomaly'] = df['z'].abs() > 1.5
df[['date','T_mean','z','is_anomaly']]
```

```python
plt.figure(figsize=(9,4))
plt.plot(df['date'], df['T_mean'], marker='o', label='T_mean')
an = df[df['is_anomaly']]
if len(an):
    plt.scatter(an['date'], an['T_mean'], color='red', s=80, label='anomaly', zorder=5)
plt.legend()
plt.title('Z-score anomalies on T_mean')
plt.ylabel('°C')
plt.tight_layout()
plt.savefig('docs/ml/metrics.png', dpi=120)
plt.show()
```

**Вывод:** Эвристика Z-score полезна для флага ручной проверки; для 14 дней лучше расширить историю. Артефакты: `docs/ml/week13_summary.md`, `anomalies_top.csv`.
